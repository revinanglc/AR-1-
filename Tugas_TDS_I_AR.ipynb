```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

np.random.seed(10)

n = 100
phi1 = 0.2
mu = 1
sigma = 1

# Fungsi simulasi AR(1)
def simulate_ar1(phi, mu, n, sigma=1.0):

    Y = np.zeros(n)
    Y[0] = mu/(1-phi)

    et = np.random.normal(0, sigma, n)

    for t in range(1,n):
        Y[t] = mu + phi*Y[t-1] + et[t]

    return Y

# Generate data
Y = simulate_ar1(phi1, mu, n)

# Plot data
plt.plot(Y)
plt.title("Simulasi Data AR(1)")
plt.show()
```

```python
model = AutoReg(Y, lags=1)
model_fit = model.fit()

print(model_fit.summary())
```
